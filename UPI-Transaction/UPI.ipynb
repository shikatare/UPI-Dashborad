```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("UPI apps transaction data in 2021.csv")
df.head()
```

```python
df.columns = df.columns.str.lower().str.strip().str.replace(" ", "_")
```

```python
df.head()
```

```python

# convert month/year into datetime
df['date'] = pd.to_datetime(df['month'].astype(str) + "-2021", format="%m-%Y")
```

```python
df.head()
```

```python
df=df.sort_values('date')
```

```python
df.head()
```

```python
df['upi_banks']=df['upi_banks'].str.capitalize()
```

```python
df.head()
```

```python
print(df.columns.tolist())
```

# Performing EDA 

```python
# App wise total transaction
top_app=df.groupby('upi_banks')['volume_(mn)'].sum().sort_values(ascending=False)
plt.figure(figsize=(15,16))
sns.barplot(x=top_app.values,y=top_app.index,palette="Accent")
plt.title("Total UPI Transaction by App in 2021")
plt.xlabel("Volume (Million)")
plt.ylabel("UPI APP")
plt.tight_layout()
plt.show()
```

```python
#Monthly Transaction Trend 
plt.figure(figsize=(14,6))
sns.lineplot(data=df,x='date',y='volume_(mn)',hue='upi_banks',marker='o')
plt.title('Monthly Transaction Trend')
plt.xlabel('Date')
plt.ylabel('Volume (mn)')
plt.legend(title="UPI App", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Monthly Transaction Value Trend
plt.figure(figsize=(15,16))
sns.lineplot(data=df,x='date',y='value_(cr)',hue='upi_banks',marker='o')
plt.title("Monthly Transaction Value by App ( Crore)")
plt.xlabel('Month')
plt.ylabel('Valur( Cr)')
plt.legend(title="UPI App", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Monthly Share of Apps
# Pivot for plotting
pivot_df = df.pivot_table(index='date', columns='upi_banks', values='volume_(mn)', aggfunc='sum')
pivot_df.fillna(0, inplace=True)

# Plot
pivot_df.plot.area(figsize=(20, 15), colormap='tab10')
plt.title("UPI Transaction Volume Share by App Over Time")
plt.ylabel("Volume (Million)")
plt.xlabel("Month")
plt.tight_layout()
plt.show()
```
